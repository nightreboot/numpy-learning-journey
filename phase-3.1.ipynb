### Filling Missing Data

### Finding Missing Data

```python
import numpy as np
import pandas as pd
```

```python
data={
    'A':[1,np.nan,3,4,5],
    'B':[1,2,3,4,5],
    'C':[1,2,3,np.nan,np.nan],
    'D':[1,np.nan,np.nan,np.nan,5]
}
df=pd.DataFrame(data)
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

### Removing Missing Data

```python
df.dropna(thresh=4)    ## thresh use for how much nan value
```

```python
df.dropna()
```

### Filling Missing Data

```python
df
```

```python
df.fillna(df.mean())
```

```python
value={'A':200,'B':100,'C':129,'D':788}
df.fillna(value=value)
```
